```python
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset("titanic")
```

```python
df.head()
```

```python
df.groupby("sex")[["survived"]].sum()
```

```python
df.groupby("sex")[["survived"]].sum()
```

```python
df.groupby(["sex", "class"])[["survived"]].sum()
```

```python
df.groupby(["sex", "class"])[["survived"]].aggregate("sum").unstack()
```

```python
df.pivot_table("survived",index="sex",columns="class")
```

```python
?df.pivot_table
```

```python
df.pivot_table("survived",index="sex",columns="class", aggfunc="sum")
```

```python
yas=pd.cut(df.age,[0,25,50,100])
```

```python
df.pivot_table("survived",["sex",yas],"class", aggfunc="sum")
```

```python
df_fl=sns.load_dataset("flights")
```

```python
df_fl.pivot_table("passengers",index="month",columns="year")
```

```python
donem=pd.cut(df_fl.year,[0,1955,2000])
```

```python
df_fl.pivot_table("passengers",["month"],donem,aggfunc="mean")
```
